```r
include <- function(pkg) {
  if (!suppressMessages(require(pkg, character.only = TRUE)))
    install.packages(pkg, character.only = TRUE)
  suppressMessages(library(pkg, pkg, character.only = TRUE))
}

include("alakazam")
include("shazam")
include("dplyr")
include("ggplot2")
```

```r
clones = read.csv(
    "/data/samples/AIRR-Seq/OURS/S3987Nr2/S3987Nr2-PBMC1_heavy/clones/S3987Nr2-PBMC1_heavy_novj_with_clones.tsv", 
    sep='\t'
)

clones = clones[order(-clones[, "duplicate_count"]), ]

print(colnames(clones))
head(clones)
```
